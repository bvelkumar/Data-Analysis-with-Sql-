```python
#To display the all details in the table of fooddata


import mysql.connector

mydb=mysql.connector.connect(host="localhost",user="root",passwd="root",database="foodsales")

mycursor = mydb.cursor()

mycursor.execute("SELECT * from fooddata")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To display the OrderDate,Region Wise from the Table of fooddata
#SELECT column1, column2, ...FROM table_name;

mycursor.execute("SELECT OrderDate,Region from fooddata")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To display the Limit 10 Table of fooddata


mycursor.execute("SELECT * from fooddata limit 10")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To display the minimum price details of the product Table of fooddata

import mysql.connector

mydb=mysql.connector.connect(host="localhost",user="root",passwd="root",database="foodsales")

mycursor = mydb.cursor()

mycursor.execute("SELECT MIN(TotalPrice) AS SmallestPrice FROM fooddata; ")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To display the Maximum price details of the product Table of fooddata


mycursor.execute("SELECT MAX(TotalPrice) AS LargestPrice FROM fooddata")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To count the number of products in the  Table of fooddata


mycursor.execute("SELECT COUNT(Product) FROM fooddata")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)

```

```python
#To dispaly the average price products in the Table of fooddata


mycursor.execute("SELECT AVG(UnitPrice) FROM fooddata")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)


```

```python
#To dispaly SUM of Total price in the Table of fooddata


mycursor.execute("SELECT SUM(TotalPrice) FROM fooddata")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)

```

```python
#To dispaly SUM of Total product in the Table of fooddata

mycursor.execute("SELECT SUM(Quantity) FROM fooddata")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)

```

```python
#To dispaly city wise customer order details using in operator in the Table of fooddata


mycursor.execute("SELECT * FROM fooddata WHERE City IN ('Boston', 'NewYork', 'LosAngeles')")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To dispaly city wise customer order details using in operator in the Table of fooddata

mycursor.execute("SELECT * FROM fooddata WHERE OrderDate IN ('1/1/2020', '9/1/2021')")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To dispaly city wise customer order details using not in operator in the Table of fooddata

mycursor.execute("SELECT * FROM fooddata WHERE OrderDate NOT IN ('1/1/2020', '9/1/2021')")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To dispaly the city wise count in group by category wise in the Table of fooddata

mycursor.execute("SELECT COUNT(OrderDate), City FROM fooddata GROUP BY Category;")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```

```python
#To dispaly the product wise count in group by city wise in the Table of fooddata

mycursor.execute("SELECT COUNT(OrderDate), Product FROM fooddata GROUP BY City;")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)

```

```python
#To display the quantity more than 50 wise orders in the Table of fooddata

mycursor.execute("SELECT COUNT(Quantity), City FROM fooddata GROUP BY City HAVING COUNT(Quantity) > 50;")

myresult=mycursor.fetchall()
for x in myresult:
    print(x)
```
